# Strict equality (===)

```python
=== 與 !== 是 no type coersion，所以盡量使用 === 與 !== 而不是 == 與 != 
```

```python
console.log(1 === 1);
// excepted output: true

console.log('hello' === 'hello');
// excepted output: true
    
console.log('1' === 1);
// excepted output: false
    
console.log(0 === false);
// excepted output: false+
```

### Description

```python
"hello" === "hello";  // true
"hello" === "hola";   // false

3 === 3; //true
3 === 4; //false

true === true  //true
true === false // false

null === null; // true
```

```python
"3" === 3;    //false
true === 1;   // false 
null === undefined;   //false
```

```python
const object1 = {
    key: "value",
};

const object2 = {
    key:"value",
};

console.log(object1 === object2); // false
console.log(object1 === object1); // true
```
